### Transform input for 1A to input for 1B

```python
import pandas as pd
```

```python
for i in ['01','03','04','07','15']:
    filename = f"../Figure1Ainput/pep_{i}.txt"
    with open(filename, "r") as f:
        lines = [line.strip() for line in f]
    result = {line.split("_")[0] for line in lines}
    with open(f"pep_{i}.txt", "w") as f:
        for item in sorted(result):
            f.write(item + "\n")
```
